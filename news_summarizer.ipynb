```python
import re

def clean_tags(text):
    clean = re.compile('<.*?>') # Compiling tags
    clean = re.sub(clean, '', text) # Replacing tags text by an empty string

    # Removing empty dialogues
    clean = '\n'.join([line for line in clean.split('\n') if not re.match('.*:\s*$', line)])

    return clean
```

```python
# Defining function to clean every text in the dataset.
def clean_df(df, cols):
    for col in cols:
        df[col] = df[col].fillna('').apply(clean_tags)
    return df
```

```python
# Cleaning texts in all datasets
train_df = clean_df(train_df,['article', 'highlights'])
test_df = clean_df(test_df,['article', 'highlights'])
validation_df = clean_df(validation_df,['article', 'highlights'])
```

```python
# Transforming dataframes into datasets

from datasets import Dataset

train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)
validation_ds = Dataset.from_pandas(validation_df)

# Visualizing results
print(train_ds)
print('\n' * 2)
print(test_ds)
print('\n' * 2)
print(validation_ds)
```

```python
# Filter out invalid rows where `article` is not a string or is empty

def filter_invalid_examples(example):
    return isinstance(example["article"], str) and len(example["article"].strip()) > 0

# Apply filter to datasets
train_ds = train_ds.filter(filter_invalid_examples)
test_ds = test_ds.filter(filter_invalid_examples)
validation_ds = validation_ds.filter(filter_invalid_examples)
```

```python
### Pre-Process the dataset
```

```python
# Define a Tokenization function

def tokenize_function(example):
    start_prompt = 'Summarize the following article.\n\n'
    end_prompt = '\n\nSummary: '
    inputs = [start_prompt + article + end_prompt for article in example["article"]]
    example['input_ids'] = tokenizer(inputs, padding="max_length", truncation=True, return_tensors="pt").input_ids
    example['labels'] = tokenizer(example["highlights"], padding="max_length", truncation=True, return_tensors="pt").input_ids

    return example

# Tokenize datasets, all the 3 splits: train, validation, ,and test in batches.
tokenized_train = train_ds.map(tokenize_function, batched=True)
tokenized_train = tokenized_train.remove_columns(['article', 'highlights'])
tokenized_test = test_ds.map(tokenize_function, batched=True)
tokenized_test = tokenized_test.remove_columns(['article', 'highlights'])
tokenized_validation = validation_ds.map(tokenize_function, batched=True)
tokenized_validation = tokenized_validation.remove_columns(['article', 'highlights'])
```

```python
print(f"Shapes of the datasets:")
print(f"Preprocessed Training Dataset: {tokenized_train}")
print(f"Preprocessed Validation Dataset: {tokenized_validation}")
print(f"Preprocessed Test Dataset: {tokenized_test}")

# Display a sample tokenized entry
print("\nSample Tokenized Train Entry:")
print(tokenized_train[0])
```

```python
tokenized_train_sample = tokenized_train.shuffle(seed=42).select(range(1000))

tokenized_val_sample = tokenized_validation.shuffle(seed=42).select(range(500))
```

```python
print(f"Shapes of the sample datasets:")
print(f"Preprocessed Training Sample Dataset: {tokenized_train_sample}")
print(f"Preprocessed Validation Sample Dataset: {tokenized_val_sample}")
```

```python
# Fully Fine-Tune the model

import os

os.environ['PYTORCH_MPS_HIGH_WATERMARK_RATIO'] = '0.0'
```

```python
from transformers import TrainingArguments, Trainer, DataCollatorForSeq2Seq

# Define training arguments

output_dir = f'./results-summary-training-{str(int(time.time()))}'

training_args = TrainingArguments(
    output_dir="output_dir",              # Where to save model and logs
    learning_rate=3e-4,                   # Learning rate for fine-tuning
    num_train_epochs=3,                   # Number of epochs
    weight_decay=0.01,                    # Weight decay for regularization
    eval_strategy="epoch",                # Evaluate after each epoch
    logging_strategy="steps",
    logging_steps=400,
    save_strategy="epoch",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    fp16=False,
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    report_to="none"
)

# Initialize Trainer
trainer = Trainer(
    model=original_model,
    args=training_args,
    train_dataset=tokenized_train_sample,
    eval_dataset=tokenized_val_sample,

)

# Train the model
trainer.train()
```

```python
from peft import LoraConfig, TaskType, get_peft_model

lora_config = LoraConfig(
    task_type=TaskType.SEQ_2_SEQ_LM,        # FLAN-T5
    r=32,                                   #Rank
    lora_alpha=32,
    target_modules=["q", "v"],
    lora_dropout=0.05,
    bias="none"
)
```

```python
lora_peft_model = get_peft_model(original_model, lora_config)
```

```python
lora_peft_model.print_trainable_parameters()
```

```python
### Train the LoRA Adapter
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer, AutoModelForSeq2SeqLM, AutoTokenizer, TrainingArguments, Trainer

output_dir = f'./results-summary-training-{str(int(time.time()))}'

lora_peft_training_args = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=20,
    auto_find_batch_size=True,
    gradient_accumulation_steps=2,
    learning_rate=1e-3,
    weight_decay=0.01,
    eval_strategy='steps',
    eval_steps=1000,
    save_strategy='steps',
    save_steps=1000,
    logging_steps=1000,
    warmup_steps=500,
    save_total_limit=3,
    load_best_model_at_end=True,
    report_to='none'
)

lora_peft_trainer = Trainer(
    model=lora_peft_model,
    args=lora_peft_training_args,
    train_dataset=tokenized_train_sample,
    eval_dataset=tokenized_val_sample,
)
```

```python
# Start training without converting to any unsupported data type

lora_peft_trainer.train()
lora_peft_model_path="./lora_peft-news-summary-checkpoint-local"
lora_peft_trainer.model.save_pretrained(lora_peft_model_path)
tokenizer.save_pretrained(lora_peft_model_path)
```
